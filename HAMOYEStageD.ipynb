```python
import pandas as pd
import numpy as np
import datetime
import matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option ('display.max_rows',500)
pd.set_option ('display.max_columns',500)
pd.set_option ('display.width',1000)
```

```python
#importing Food Production Dataset
Food_Production = pd.read_csv("C:/Users/User/Desktop/Hamoye Internship/Africa Food Production (2004 - 2013).csv")
```

```python
Food_Production
```

```python
Food_Production.info()
```

```python
#converting dates to Datetime
pd.to_datetime(Food_Production['Year'], format ="%Y").dt.year
```

```python
#Grouping Food Production Dataset by Country, Year and then by item
Group =Food_Production.groupby(['Country', 'Year','Item'])
```

```python
Group.first()
```

```python
#Grouping by Country and summing the values of all items for each year
Totalled = Food_Production.groupby(['Country', 'Year'])['Value'].sum()
```

```python
Totalled
```

```python
#Visualizing Dataset to show the trend in food production between 2004 and 2013
sns.set_style('dark')
Plot = Food_Production.groupby('Country')['Value'].sum().sort_values (ascending=False)
plt.ylabel('Value (In Log Scale)')
plt.title('Total Food Production by Country')
Plot.plot(kind='bar', figsize=[10,5],logy=True)
```
